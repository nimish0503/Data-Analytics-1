# Assignment 2: Central Limit Theorem and Sampling Methods

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data Loading and Overview
# Loading Iris Dataset
iris = sns.load_dataset('iris')
print("First few rows of the dataset:")
print(iris.head())
```

```python
# Simple Random Sampling
# 30 Observations (Random Sample)
random_sample = iris.sample(n=30, random_state=42)
print("Simple Random Sample:")
print(random_sample)
```

```python
# 3. Sample Mean Distribution Analysis
# Repeat random sampling 100 times, calculate sample means for 'sepal_length'
sample_means = []
for _ in range(100):
  sample = iris['sepal_length'].sample(n=30, random_state=np.random.randint(0, 1000))
  sample_means.append(sample.mean())
```

```python
# Ploting Distribution of Sample Means
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=15, color='skyblue', edgecolor='black')
plt.title("Distribution of Sample Means (Sepal Length)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Systemic Sampling
def systematic_sampling(data, step):
    indices = np.arange(0, len(data), step)
    return data.iloc[indices]
```

```python
# Systemic Sample(20% of the dataset)
systemic_sample = iris.sample(frac=0.2, random_state=42)
print("Systematic Sample:")
print(systematic_sample)
```

```python
# Plot original dataset and sample dataset
plt.figure(figsize=(8, 5))
plt.scatter(range(len(iris)), iris['sepal_length'], color='gray', alpha=0.5, label='Original Dataset')
plt.scatter(systematic_sample.index, systematic_sample['sepal_length'], color='red', label='Systematic Sample')
plt.title("Systematic Sampling: Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.legend()
plt.show()
```

# Key Insights from the Analysis
- The Central Limit Theorem is demonstrated effectively, with the distribution of sample means forming a normal-like curve.
- Random sampling ensures unbiased subsets of the data, which can be used to generalize findings to the entire dataset.
- Sampling methods like systematic or random can both provide smaller datasets, but random sampling is less prone to periodic biases.
