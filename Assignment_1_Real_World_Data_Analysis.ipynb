# Assignment 1: Real World Data Analysis
The dataset is the World Happiness Report 2024, which evaluates global happiness based on metrics like "Ladder score," GDP per capita, social support, and more.
### A. Descriptive Analysis
1. Measures of Central Tendency -
Mean, Median and Mode
2. Dispersion -
Variance, Standard Deviation and Range

### B. Graphs Used
1. Scatter Plot with Regression Line
2. Box Plot
3. Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load the dataset
file_path = '/content/drive/MyDrive/DA 1 Assignments/World-happiness-report-2024.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Computing statistical measures
Mean = data['Healthy life expectancy'].mean()
Median = data['Healthy life expectancy'].median()
Mode = data['Healthy life expectancy'].mode().iloc[0]
Variance = data['Healthy life expectancy'].var()
Standard_Deviation = data['Healthy life expectancy'].std()
Range = data['Healthy life expectancy'].max() - data['Healthy life expectancy'].min()
print("Mean: ", Mean)
print("Median: ", Median)
print("Mode: ", Mode)
print("Variance: ", Variance)
print("Standard Deviation: ", Standard_Deviation)
print("Range: ", Range)
```

```python
# scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Healthy life expectancy', y='Ladder score', data=data)
plt.title('Relationship between Healthy Life Expectancy and Ladder Score')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Ladder Score')
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Regional indicator', y='Ladder score', data=data)
plt.title('Ladder Score Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Ladder Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['Ladder score'], kde=True)
plt.title('Distribution of Ladder Score')
plt.xlabel('Ladder Score')
plt.ylabel('Frequency')
plt.show()
```
